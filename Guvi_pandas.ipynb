```

def isPrime(n):
	

	if n <= 1 :
		return False

	
	for i in range(2, n):
		if n % i == 0:
			return False

	return True


def printPrime(n):
	for i in range(2, n + 1):
		if isPrime(i):
			print(i, end = " ")


if __name__ == "__main__" :
	n = 7

	printPrime(n)

```

```
def primenumber(MyNum):
  n = 0
  i = 2
  for i in range(2,MyNum//2+1):
    if MyNum % i == 0:
      n = n + 1
      break
  if n == 0:
    print(MyNum, end=" ")

x =int(input())
print("Prime numbers less than", x, "are:")
for i in range(2, x+1):
  primenumber(i)
```
